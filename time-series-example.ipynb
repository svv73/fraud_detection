```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df = pd.read_csv('ts_1.csv', parse_dates=["Date"])
```

```python
df
```

```python
x = df['Date'].values
y = df['Value'].values
plt.plot(x, y)
plt.xlabel('Дата транзакции')
plt.ylabel('Сумма транзакции')
plt.show()
```

```python
result_mul = seasonal_decompose(x=df['Value'], model='additive', extrapolate_trend='freq', period=8)
plt.rcParams.update({'figure.figsize': (20,15)})
result_mul.plot()
plt.show()
```
